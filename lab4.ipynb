### Exercício 1: Normal por Aceitação-Rejeição


```python
def bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
def exponential(lam=1):
    u = np.random.uniform()
    x = - np.log(u) / lam
    return x
```

```python
def my_normal(a=0, b=1):
    u = np.random.uniform()
    y = exponential()
    while u > np.exp(-((y-1) ** 2)/2):
        u = np.random.uniform()
        y = exponential()
    if bernoulli(0.5):
        y *= -1
    return a + np.sqrt(b) * y
```

```python
my_normal_trials = [my_normal(0, 1) for i in range(100000)]
plt.hist(my_normal_trials)
plt.show()
```

```python
np_normal_trials = [np.random.normal(0, 1) for i in range(1000000)]
plt.hist(np_normal_trials)
plt.show()
```
