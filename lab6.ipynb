### Distribuições

```python
def my_bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
def my_exponential(lam=1):
    u = np.random.uniform()
    x = - np.log(u) / lam
    return x
```

```python
def my_normal(a=0, b=1):
    u = np.random.uniform()
    y = my_exponential()
    while u > np.exp(-((y-1) ** 2)/2):
        u = np.random.uniform()
        y = my_exponential()
    if my_bernoulli(0.5):
        y *= -1
    return a + np.sqrt(b) * y
```

### Exemplo 9

```python
sum = 0
sum_of_sqr = 0
n = 10000
for i in range(n):
    X = np.random.uniform(0, 1)
    gx = np.exp(-X**2)
    sum += gx
    sum_of_sqr += gx ** 2

mean = sum / n
mean_of_sqr = sum_of_sqr / n
s2 = (mean_of_sqr - mean ** 2) * n / (n-1)
sd = s2 ** 0.5

std_error = sd / (10000 ** 0.5)

print(f"Mean = {mean:.2f}\n95% confidence interval: ({mean - std_error * 2:.3f}, {mean + std_error * 2:.3f})")
```

### Verificando o intervalo de confiança

```python
def pi_CI():
    sample = []
    n = 10000
    for i in range(n):
        X = np.random.uniform()
        Y = np.random.uniform()
        if X ** 2 + Y ** 2 <= 1:
            sample.append(1)
        else:
            sample.append(0)
    pi = 4 * np.mean(sample)
    sd = 4 * np.std(sample)
    std_error = sd / n ** 0.5
    return [pi - std_error * 2, pi + std_error * 2]
```

```python
def check_if_pi_is_within_CI(CI:list):
    if np.pi >= CI[0] and np.pi <= CI[1]:
        return 1
    return 0
```

```python
pi_within_CI_sample = []
for i in range(1000):
    pi_within_CI_sample.append(check_if_pi_is_within_CI(pi_CI()))
np.mean(pi_within_CI_sample)
```
