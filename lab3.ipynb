## Exerc√≠cio 1: Exponencial

```python
def exponential(lam):
    U = np.random.uniform(0, 1)
    x = (-1/lam) * np.log(1 - U)
    return x
```

```python
my_exponential = [exponential(3) for i in range(10000)]
my_exponential
```

```python
plt.hist(my_exponential)
```

```python
np_exponential = [np.random.exponential(1/3) for i in range(10000)]
plt.hist(np_exponential)
```

```python
my_exponential = [exponential(0.5) for i in range(10000)]
plt.hist(my_exponential)
```

```python
my_exponential = [exponential(5) for i in range(10000)]
plt.hist(my_exponential)
```

Valores maiores para lambda concentram a distribui√ß√£o mais pr√≥ximo de 0. Valores menores "esticam" a distribui√ß√£o, distanciando-a do 0. Isso faz sentido, pois se lambda representa o n√∫mero m√©dio de eventos de poisson em uma unidade de tempo, conforme lambda cresce, o tempo entre eventos diminui.

## Exerc√≠cio 2: Aproxima√ß√£o da Exponencial pela Geom√©trica

```python
def geometric(p):
    unif = np.random.uniform(0, 1)
    x = math.floor(np.log(unif)/np.log(1-p)) + 1
    return x
```

```python
def exponential_from_geometric(lam, n):
    geom = geometric(lam/n)
    x = geom/n
    return x
```

```python
def density_of_exponential(x, lam=1):
    return np.exp(-lam*x)*lam
```

```python
exponential_n50 = [exponential_from_geometric(1, 50) for i in range(10000)]
exponential_n500 = [exponential_from_geometric(1, 500) for i in range(10000)]
exponential_n5000 = [exponential_from_geometric(1, 5000) for i in range(10000)]
```

```python
x = np.linspace(0, 5, 1000)
plt.plot(x, density_of_exponential(x))
plt.hist(exponential_n50, density=True)
```

```python
x = np.linspace(0, 5, 1000)
plt.plot(x, density_of_exponential(x))
plt.hist(exponential_n500, density=True)
```

```python
x = np.linspace(0, 5, 1000)
plt.plot(x, density_of_exponential(x))
plt.hist(exponential_n5000, density=True)
```

## Exerc√≠cio 3: Poisson a partir de Exponenciais

```python
def poisson_via_exp(mu):
    t = 0
    x = 0
    while t < 1:
        t += exponential(mu)
        if t < 1:
            x += 1
    return x
```

```python
lam = 4
poisson_via_exp_trials = [poisson_via_exp(lam) for i in range(10000)]
plt.hist(poisson_via_exp_trials)
plt.show()
```

```python
poisson_numpy_trials = [np.random.poisson(4) for i in range(10000)]
plt.hist(poisson_numpy_trials)
plt.show()
```

```python
poisson_mean = np.mean(poisson_via_exp_trials)
poisson_var = np.var(poisson_via_exp_trials)
print(f"M√©dia te√≥rica: {lam}       | Vari√¢ncia te√≥rica: {lam}\n"
      f"M√©dia emp√≠rica: {poisson_mean:.3f} | Vari√¢ncia emp√≠rica: {poisson_var}")
```
