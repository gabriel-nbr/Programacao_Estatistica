```python
import numpy as np
import matplotlib.pyplot as plt
```

### Distribuições

```python
def my_bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
def my_exponential(lam=1):
    u = np.random.uniform()
    x = - np.log(u) / lam
    return x
```

```python
def my_normal(a=0, b=1):
    u = np.random.uniform()
    y = my_exponential()
    while u > np.exp(-((y-1) ** 2)/2):
        u = np.random.uniform()
        y = my_exponential()
    if my_bernoulli(0.5):
        y *= -1
    return a + np.sqrt(b) * y
```

### Exercício 1

```python
mean_sample_dir1 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        x = np.random.uniform(1, 2)
        sample.append(np.exp(x))
    mean_sample_dir1.append(np.mean(sample))
var_est_direta = np.var(mean_sample_dir1)
print(f"Variância da estimativa direta: {var_est_direta}")
plt.hist(mean_sample_dir1)
plt.title("Distribuição das Estimativas: Método Direto")
plt.show()
```

```python
mean_sample_ant1 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        x1 = np.random.uniform(1, 2)
        x2 = abs(x1-3)
        sample.append((np.exp(x1)+np.exp(x2))/2)
    mean_sample_ant1.append(np.mean(sample))
var_est_antt = np.var(mean_sample_ant1)
print(f"Variância da estimativa com variável antitética): {var_est_antt}")
plt.hist(mean_sample_ant1)
plt.title("Distribuição das Estimativas: Método com Variável Antitética")
plt.show()
```

```python
plt.hist(mean_sample_dir1)
plt.hist(mean_sample_ant1)
plt.title("Método Direto (Azul) vs Variável Antitética (Laranja)")
plt.show()
```

```python
print(f"A razão entre as variâncias do método com variável antitética e do método usual é: {var_est_antt/var_est_direta:.3f}")
```

### Exercício 2

```python
mean_sample_dir2 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        u = np.random.uniform()
        x = - np.log(u)
        sample.append(np.exp(- x ** 2))
    mean_sample_dir2.append(np.mean(sample))
var_est_direta = np.var(mean_sample_dir2)
print(f"Variância da estimativa direta: {var_est_direta}")
plt.hist(mean_sample_dir2)
plt.title("Distribuição das Estimativas: Método Direto")
plt.show()
```

```python
mean_sample_ant2 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        u = np.random.uniform()
        x1 = - np.log(u)
        x2 = - np.log(1-u)
        sample.append((np.exp(- x1 ** 2) + np.exp(- x2 ** 2)) / 2)
    mean_sample_ant2.append(np.mean(sample))
var_est_antt = np.var(mean_sample_ant2)
print(f"Variância da estimativa com variável antitética): {var_est_antt}")
plt.hist(mean_sample_ant2)
plt.title("Distribuição das Estimativas: Método com Variável Antitética")
plt.show()
```

```python
plt.hist(mean_sample_dir2)
plt.hist(mean_sample_ant2)
plt.title("Método Direto (Azul) vs Variável Antitética (Laranja)")
plt.show()
```

```python
print(f"A razão entre as variâncias do método com variável antitética e do método usual é: {var_est_antt/var_est_direta:.3f}")
```

```python
mean_sample_dir3 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        x = my_normal(1, 2)
        sample.append(np.exp((- x ** 2) + 1))
    mean_sample_dir3.append(np.mean(sample))
var_est_direta = np.var(mean_sample_dir3)
print(f"Variância da estimativa direta: {var_est_direta}")
plt.hist(mean_sample_dir3)
plt.title("Distribuição das Estimativas: Método Direto")
plt.show()
```

```python
mean_sample_ant3 = []
m = 1000
for i in range(m):
    sample = []
    n = 1000
    for j in range(n):
        u = np.random.uniform()
        x = my_normal(1, 2)
        sample.append((np.exp(- x ** 2 + 1) + np.exp(- (2 - x) ** 2 + 1)) / 2)
    mean_sample_ant3.append(np.mean(sample))
var_est_direta = np.var(mean_sample_ant3)
print(f"Variância da estimativa direta: {var_est_direta}")
plt.hist(mean_sample_ant3)
plt.title("Distribuição das Estimativas: Método com Variável Antitética")
plt.show()
```
