### Exercício 3: Produto interno e distância

```python
def dot_product(vector1, vector2):
    if len(vector1) != len(vector2):
        raise IndexError("Vetores precisam ter o mesmo tamanho!")
    
    dot_product = 0
    for i in range(len(vector1)):
        dot_product += vector1[i] * vector2[i]
    
    return dot_product
```

```python
vetor1 = np.random.randint(0, 5, size=5)
vetor2 = np.random.randint(0, 5, size=5)
print(f"Vetor 1: {vetor1}")
print(f"Vetor 2: {vetor2}")
print(f"Meu produto escalar: {dot_product(vetor1, vetor2)}")
print(f"Produto escalar np: {np.dot(vetor1, vetor2)}")
```

```python
def euclidean_distance(vector1, vector2):
    if len(vector1) != len(vector2):
        raise IndexError("Vetores precisam ter o mesmo tamanho!")
    
    distance_vector = vector1 - vector2
    distance = dot_product(distance_vector, distance_vector) ** 0.5

    return distance
```

```python
print(f"Distância euclidiana entre os vetores (minha função): {euclidean_distance(vetor1, vetor2):.3f}")
print(f"Distância euclidiana entre os vetores (numpy): {np.linalg.norm((vetor1 - vetor2)):.3f}")
```

### Exercício 4: Matriz invertível

```python
A = np.random.randint(-5, 5, size=(4, 4))
A
```

```python
def invert_matrix(matrix):
    if np.linalg.det(matrix) != 0:
        inverted_matrix = np.linalg.inv(matrix)
        return inverted_matrix
    else:
        return None
    
A_inv = invert_matrix(A)
if A_inv is None:
    print("A matriz não é inversível")
else:
    print(f"A * A^-1 =\n{A @ A_inv}")
```

```python
print(f"A * A^-1 =\n{np.round(A @ A_inv)}")
```

Conclusão: ``A * A^-1`` é **próximo** da matriz identidade `I`, mas não **é exatamente igual**, devido provavelmente a **erros de arredondamento** do computador

### Exercício 5: Autovalores e autovetores

```python
M = np.random.randint(0, 5, size=(4, 4))
M
```

```python
Mt = np.transpose(M)
Mt
```

```python
B = 0.5 * M @ Mt
B
```

```python
autovalores, autovetores = np.linalg.eig(B)
for i in range(len(autovalores)):
    print(f"Autovalor: {autovalores[i]}")
    print(f"Autovetor associado: {autovetores[:, i]}\n")
```

```python
for i in range(len(autovalores)):
    print(f"Autovalor * autovetor = {autovalores[i] * autovetores[:, i]}")
    print(f"Matriz B * autovetor =  {B @ autovetores[:, i]}\n")
```

### Exercício 6: Sistema linear

```python
A = np.random.uniform(-1000, 1000, size=(3, 3))
y = np.random.randint(-1000, 1000, 3)
print(A)
print(y)
```

```python
x = np.linalg.solve(A, y)
x
```

```python
result = A @ x
np.linalg.norm(result - y) ** 2
```

```python
np.round(np.linalg.norm(result - y) ** 2, 10)
```

Conclusão: A solução encontrada **não é exatamente a correta**, de novo provavelmente devido a **erros de arredondamento** do computador.
