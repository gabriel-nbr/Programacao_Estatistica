### Distribuições

```python
def my_bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
def my_exponential(lam=1):
    u = np.random.uniform()
    x = - np.log(u) / lam
    return x
```

```python
def my_normal(a=0, b=1):
    u = np.random.uniform()
    y = my_exponential()
    while u > np.exp(-((y-1) ** 2)/2):
        u = np.random.uniform()
        y = my_exponential()
    if my_bernoulli(0.5):
        y *= -1
    return a + np.sqrt(b) * y
```

### Exercício 1

```python
def pi_monte_carlo_direto(n):
    sample = []
    for i in range(n):
        x = np.random.uniform()
        fx = 1 / (1 + x ** 2)
        sample.append(fx)
    return np.mean(sample), np.var(sample)
```

```python
def pi_monte_carlo_controle(n):
    for i in range(n):
        U = np.random.uniform(size=n)

        X_sample = 1 / (1 + U ** 2)
        Y_sample = U ** 2
        Y_mean = 1/3

    XY_cov = np.cov(X_sample, Y_sample)
    c = -XY_cov[0, 1] / XY_cov[1, 1]

    Z_sample = X_sample + c * (Y_sample - Y_mean)
    return np.mean(Z_sample), np.var(Z_sample)
```

```python
print(f"O valor estimado para pi pelo método direto é: {4 * pi_monte_carlo_direto(10000)[0]:.4f}\n"
f"A variância das estimativas foi {pi_monte_carlo_direto(10000)[1]:.4f}"
f"O erro da estimativa foi {np.pi - 4 * pi_monte_carlo_direto(10000)[0]:.5f}")
```

```python
print(f"O valor estimado para pi pelo método direto é: {4 * pi_monte_carlo_controle(10000)[0]:.4f}\n"
f"A variância das estimativas foi {pi_monte_carlo_controle(10000)[1]:.4f}"
f"O erro da estimativa foi {np.pi - 4 * pi_monte_carlo_controle(10000)[0]:.5f}")
```
