```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
distribuicoes = {
    "Normal": {"params": {"loc": 5, "scale": 2}, "dados": None, "estatisticas": None},
    "Exponential": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniform": {"params": {"low": 0, "high": 10}, "dados": None, "estatisticas": None},
    "Poisson": {"params": {"lam": 5}, "dados": None, "estatisticas": None},
    "Binomial": {"params": {"n": 10, "p": 0.5}, "dados": None, "estatisticas": None}
    }

distribuicoes
```

```python
distribuicoes['Exponential']["params"]["scale"] = 1
distribuicoes
```

```python
distribuicoes['Normal']["dados"] = np.random.normal(loc=5, scale=2, size=1000)
```

```python
distribuicoes['Exponential']["dados"] = np.random.exponential(scale=1, size=1000)
```

```python
distribuicoes['Uniform']["dados"] = np.random.uniform(low=0, high=10, size=1000)
```

```python
distribuicoes['Poisson']["dados"] = np.random.poisson(lam=5, size=1000)
```

```python
distribuicoes['Binomial']["dados"] = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
distribuicoes
```

```python
def calcular_estatisticas(data):
    return {"media": np.mean(data),
            "mediana": np.median(data),
            "variancia": np.var(data),
            "desvio_padrao": np.std(data),
            "minimo": np.min(data),
            "maximo": np.max(data),
            "quartil": np.percentile(data, [25, 50, 75])
            }
```

```python
for key in distribuicoes.keys():
    distribuicoes[key]["estatisticas"] = calcular_estatisticas(distribuicoes[key]["dados"])
```

```python
plt.hist(distribuicoes["Normal"]["dados"], bins=50)
```

```python
plt.hist(distribuicoes["Exponential"]["dados"], bins=50)
```

```python
plt.hist(distribuicoes["Uniform"]["dados"], bins=20)
```

```python
plt.hist(distribuicoes["Poisson"]["dados"], bins=50)
```

```python
plt.hist(distribuicoes["Binomial"]["dados"], bins=50)
```

```python
boxplot_data = [distribuicoes[i]["dados"] for i in distribuicoes]
plt.boxplot(boxplot_data, labels=distribuicoes.keys())
```

```python
valores_teoricos = {
    "Normal": {
        "media": distribuicoes['Normal']["params"]["loc"],
        "variancia": distribuicoes['Normal']["params"]["scale"] ** 2},
    "Exponential": {
        "media": distribuicoes['Exponential']["params"]["scale"],
        "variancia": distribuicoes['Exponential']["params"]["scale"] ** 2},
    "Uniform": {
        "media": (distribuicoes['Uniform']["params"]["low"] + distribuicoes['Uniform']["params"]["high"]) / 2,
        "variancia": (distribuicoes['Uniform']["params"]["high"] - distribuicoes['Uniform']["params"]["low"]) ** 2 / 12},
    "Poisson": {
        "media": distribuicoes['Poisson']["params"]["lam"],
        "variancia": distribuicoes['Poisson']["params"]["lam"]},
    "Binomial": {
        "media": distribuicoes['Binomial']["params"]["n"] * distribuicoes['Binomial']["params"]["p"],
        "variancia": (
            distribuicoes['Binomial']["params"]["n"] * distribuicoes['Binomial']["params"]["p"] * (1 - distribuicoes['Binomial']["params"]["p"]))}
}
```

```python
for distribution, properties in distribuicoes.items():
    print(f"{distribution}: Mdia terica = {valores_teoricos[distribution]['media']:.2f} -- Mdia observada = {properties['estatisticas']['media']:.2f} \n"
    f"        Varincia terica = {valores_teoricos[distribution]['variancia']:.2f} -- Varincia observada = {properties['estatisticas']['variancia']:.2f} \n")
```

```python
barplot_media_teorica = [valores_teoricos[i]['media'] for i in distribuicoes]
plt.bar(distribuicoes.keys(), barplot_media_teorica)
```

```python
barplot_media_observada = [distribuicoes[i]['estatisticas']['media'] for i in distribuicoes]
plt.bar(distribuicoes.keys(), barplot_media_observada)
```
