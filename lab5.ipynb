### Distribuições

```python
def my_bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
def my_exponential(lam=1):
    u = np.random.uniform()
    x = - np.log(u) / lam
    return x
```

```python
def my_normal(a=0, b=1):
    u = np.random.uniform()
    y = my_exponential()
    while u > np.exp(-((y-1) ** 2)/2):
        u = np.random.uniform()
        y = my_exponential()
    if my_bernoulli(0.5):
        y *= -1
    return a + np.sqrt(b) * y
```

```python
sum = 0
n = 10000
for i in range(n):
    X = np.random.uniform(0, 1)
    sum += np.exp(-X**2)
result = sum/n
print(f"Resultado: {result:.4f}")
```

### Exemplo 11

```python
sum = 0
n = 100000
for i in range(n):
    X = np.random.uniform()
    Y = np.random.uniform()
    if X ** 2 + Y ** 2 <= 1:
        sum += 1
pi = 4 * sum / n
print(f"pi = {pi:.4f}")
```

### Exemplo 12

```python
sum = 0
n = 10000
for i in range(n):
    X = np.random.uniform()
    sum += X
mean = sum / n

sum = 0
for i in range(n):
    X = np.random.uniform()
    sum += (X - mean) ** 2
var = sum / (n - 1)
print(f"Mean = {mean:.4f}\nVar = {var:.4f}")
```
