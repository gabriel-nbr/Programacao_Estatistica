```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def bernoulli(p):
    uniform = np.random.uniform(0, 1)
    if uniform <= p:
        return 1
    else:
        return 0
```

```python
my_bernoulli_trials = [bernoulli(0.3) for i in range(50)]
numpy_bernoulli_trials = [np.random.binomial(1, 0.3) for i in range(50)]
plt.hist(my_bernoulli_trials)
```

```python
plt.hist(numpy_bernoulli_trials)
```

```python
def uniform(precision):
    result = 0
    for k in range(1, precision):
        result += bernoulli(0.5) / (2 ** k)
    return result
```

```python
my_uniform_trials = [uniform(20) for i in range(1000)]
numpy_uniform_trials = [np.random.uniform() for i in range(1000)]
plt.hist(my_uniform_trials)
```

```python
plt.hist(numpy_uniform_trials)
```

```python
def binomial_soma_de_B(n, p):
    result = 0
    for i in range(n):
        if bernoulli(p) == 1:
            result += 1
    return result
```

```python
start = time.time()
binomial_trials1 = [binomial_soma_de_B(20, 0.5) for i in range(10000)]
end = time.time()
print(f"Tempo para binomial por soma de Bernoullis: {end-start}")
plt.hist(binomial_trials1)
```

```python
def binomial_inversao_recursiva(n, p):
    unif = np.random.uniform()
    i = 0
    pi = (1-p) ** n
    F = pi
    while unif > F:
        pi = ((n - i) / (i + 1)) * p / (1-p) * pi
        i += 1
        F = F + pi
    return i
```

```python
start = time.time()
binomial_trials2 = [binomial_inversao_recursiva(20, 0.5) for i in range(10000)]
end = time.time()
print(f"Tempo para binomial por inverso recursiva: {end-start}")
plt.hist(binomial_trials1)
```
